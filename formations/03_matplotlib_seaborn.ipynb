# Affichage des points dans un repère

```python

plt.scatter(x, y)
plt.ylim(0, 16) # Redimmensionnement de l'axe Y
plt.xlim(0, 5) # Redimmensionnement de l'axe X
plt.title("Affichage des points dans un repère")
plt.ylabel("Axe des ordonnées")
plt.xlabel("Axe des abscisses")
```

```python
fig, ax = plt.subplots(figsize = (10, 8), nrows = 2, ncols = 1)
ax[0].plot(x, y)
ax[1].scatter(np.array(x) + 1, y)
```

# Création des graphe avec la méthode orientée object

```python
fig = plt.figure(figsize=(16, 8)) # Créer une figure
ax1 = fig.add_subplot(211) # Définir le nombre de subplot et créer les axes
ax2 = fig.add_subplot(212)

# Modifier chaque axe pour faire l'affichage
ax1.set_xlabel("X")
ax1.set_ylabel("Y")
ax1.plot(x, y)


ax2.set_xlabel("Catégories")
ax2.set_ylabel("Tailles")
categories = ["Hommes", "Femmes"]
tailles = [180, 160]
ax2.bar(categories, tailles)
```

```python
notes = pd.read_csv("students_notes.csv", delimiter=';')
notes
```

```python
notes.plot(kind="scatter", figsize=[10, 5], x="maths", y="anglais", color="black")
```

```python
# Affichage d'un histogramme
notes.plot(kind="hist", x="maths", y="anglais", xlabel="Notes de maths", ylabel="Notes d'anglais")
```

```python
# Histogramme des notes de mathématiques juste
maths  = notes[(notes.maths < 5)]
maths["maths"].plot(kind="hist", figsize=[10, 5], xlabel="Notes de maths", ylabel="Nombre d'étudiants")
```

```python
notes[(notes.maths >= 3.5) & (notes.maths <= 4.0)]["maths"].count() # Nombre d'étudiants ayant eu une note de maths comprise en [3.5, 4.0]
```

```python
notes["maths"].value_counts().plot.pie(legend = False, figsize = [7, 7])
```

```python
notes["maths"].value_counts(ascending=False).plot(
    figsize=(14, 6),
    rot=0,
    color='brown',
    kind="bar",
    ylabel="Nombres d'étudiants",
    xlabel="Notes des d'étudiants",
    ylim=(0, 60),
    title="Nombres d'étudiants par notes des d'étudiants",
    
)
```
